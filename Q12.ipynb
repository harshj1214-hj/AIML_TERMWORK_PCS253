```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
# READ DATASET 
df = pd.read_csv("disease_risk.csv")

print("Dataset Shape:", df.shape)
print(df.head())
```

```python
# DATA PREPROCESSING
df = df.drop('patient_id', axis=1)

# Convert categorical to numeric
df = pd.get_dummies(df, drop_first=True)

# Handle missing values
df = df.fillna(df.mean())
```

```python
y = df["Disease_Risk"]               # Target
X = df.drop("Disease_Risk", axis=1)  # Features
```

```python
# TRAIN-TEST SPLIT (70:30)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# FEATURE SCALING
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# LOGISTIC REGRESSION MODEL
model = LogisticRegression()

model.fit(X_train_scaled, y_train)
```

```python
coefficients = pd.DataFrame({
    "Feature": X.columns,
    "Coefficient": model.coef_[0]
})

print(coefficients)
print("Intercept:", model.intercept_[0])
```

```python
# MODEL EVALUATION

y_pred = model.predict(X_test_scaled)

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print("\nAccuracy:", accuracy)

# Precision
precision=precision_score(y_test, y_pred)
print("\nprecision", precision)

# Recall
recall=recall_score(y_test, y_pred)
print("\nrecall", recall)

# F1_score
f1=f1_score(y_test, y_pred)
print("\nf1_score", f1)

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:\n", cm)


print("\nclassification_report",classification_report(y_test, y_pred))
```

```python
# PREDICT NEW DATA
new_data = [[45, 120, 200, 90, 75, 1, 3,9]]  # example values

new_data_scaled = scaler.transform(new_data)

prediction = model.predict(new_data_scaled)

print("\nPrediction for new data (0=No Risk, 1=Risk):", prediction[0])
```

```python
# ANALYSIS
# ============================
print("\n--- Analysis ---")

print("Model coefficients show how each feature affects disease risk.")
print("Positive coefficient increases risk, negative decreases risk.")
print("Accuracy indicates overall performance of the model.")
print("Confusion matrix shows correct and incorrect predictions.")
```
