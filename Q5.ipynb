```python
#bfs traversal
from collections import deque
def bfs (st,goal,graph):
    queue=deque([st])
    vis=set([st])
    path=[]
    while queue:
        node=queue.popleft()
        path.append(node)
        if node == goal:
            print(path)
            return True
        if node not in vis:
            vis.add(node)
        for child in graph[node]:
            if child not in vis:
                queue.append(child)
    return False
```

```python
# shortest path 

from collections import deque

def short(st, goal, graph):
    queue = deque([[st]])
    vis = set([st])

    while queue:
        path = queue.popleft()
        node = path[-1]

        if node == goal:
            return path

        for child in graph[node]:
            if child not in vis:
                vis.add(child)
                new_path = path + [child]
                queue.append(new_path)

    return None
```

```python
# Static Graph
graph_static = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E', 'F'],
    'C': ['A', 'D', 'F'],
    'D': ['B', 'C', 'E', 'F'],
    'E': ['B', 'D', 'F'],
    'F': ['C', 'D', 'E', 'B']
}

print("Static Graph:", graph_static)
```

```python
# Input start & goal
start = input("Enter start node (Static): ").strip().upper()
goal = input("Enter goal node (Static): ").strip().upper()

path=bfs (start,goal,graph_static)

if path:
    print("BFS Path (Static):", path)
else:
    print("No path found")
```

```python
# Input start & goal for shortest path
start = input("Enter start node (Static): ").strip().upper()
goal = input("Enter goal node (Static): ").strip().upper()

short_path=short (start,goal,graph_static)

if path:
    print("BFS shortest Path (Static):",short_path)
else:
    print("No path found")
```

```python
# Dynamic Graph 

n = int(input("Enter number of nodes: "))
e = int(input("Enter number of edges: "))
d=int(input("graph directed or not ( 1 for non directed and 0 for directed)"))
graph = {}
for i in range(n):
    node = input("Enter node name: ").strip().upper()
    graph[node] = []

for i in range(e):
    u = input("Enter starting node: ").strip().upper()
    v = input("Enter ending node: ").strip().upper()
    
    graph[u].append(v)
    if d:
        graph[v].append(u)   

print("Graph =", graph)
```

```python
# Input start & goal
start = input("Enter start node (Dynamic): ").strip().upper()
goal = input("Enter goal node (Dynamic): ").strip().upper()

path=bfs (start,goal,graph)

if path:
    print("BFS Path (Dynamic):", path)
else:
    print("No path found")
```
