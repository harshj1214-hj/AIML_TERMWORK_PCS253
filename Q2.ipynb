```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create Dataset

data = {
    "height": [1.70, 1.65, 1.80, 1.75, 1.60, 1.68, 1.72, 1.66, np.nan, 1.74],
    "weight": [65, 55, 75, 70, 50, 60, 68, 58, 62, 72],
    "gender": ["M", "F", "M", "F", None, "M", "F", "F", "M", None]
}

df = pd.DataFrame(data)
print(df)
```

```python
# i. Compute BMI
df["BMI"] = df["weight"] / (df["height"] ** 2)
print("Dataset with BMI:\n", df)
```

```python
# Save to CSV
df.to_csv("bmi_data.csv", index=False)
```

```python
#Load CSV
df = pd.read_csv("bmi_data.csv")
print(df)
```

```python
print("Shape:", df.shape)
print("Size:", df.size)
```

```python
print("Null in Gender:", df["gender"].isnull().sum())
```

```python
# Remove null from Gender
df = df.dropna(subset=["gender"])
```

```python
print("\nInfo:\n")
print(df.info())

print("\nDescription:\n")
print(df.describe())
```

```python
#  Histogram on Gender
df["gender"].value_counts().plot(kind='bar')
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
sns.scatterplot(x=df["height"], y=df["BMI"], hue=df["gender"])
plt.title("Height vs BMI")
plt.show()

print("Inference: BMI generally increases with weight and varies slightly with height.")
```

```python
print("\nMean:\n", df[["height","weight","BMI"]].mean())
print("\nMax:\n", df[["height","weight","BMI"]].max())
print("\nMin:\n", df[["height","weight","BMI"]].min())
print("\nSum:\n", df[["height","weight","BMI"]].sum())
```

```python
print("\nBMI less than 20:\n")
print(df[df["BMI"] < 20])
```
