```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
#  Load dataset

df = pd.read_csv("healthcare.csv")
print(df)
```

```python
print(df.head())
```

```python
#  Min-Max Scaling
scaler = MinMaxScaler()
df_minmax = df.copy()
numeric_cols = ["age", "height_cm", "weight_kg", "blood_pressure", "sugar_level"]
df_minmax[numeric_cols] = scaler.fit_transform(df[numeric_cols])
```

```python
# Plot graphs (Min-Max)
df_minmax[numeric_cols].hist(figsize=(10,6))
plt.suptitle("Min-Max Normalized Data")
plt.show()
```

```python

#  Standard Scaling

standard = StandardScaler()
df_standard = df.copy()
df_standard[numeric_cols] = standard.fit_transform(df[numeric_cols])
```

```python
# Plot graphs (Standard Scaler)
df_standard[numeric_cols].hist(figsize=(10,6))
plt.suptitle("Standard Scaled Data")
plt.show()
```

```python
for col in numeric_cols:
    
    plt.plot(df[col], label="Original")
    plt.plot(df_minmax[col], label="MinMax")
    plt.plot(df_standard[col], label="Standard")
    
    plt.title(col)
    plt.legend()
    plt.show()
```
