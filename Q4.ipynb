```python
# DFS Function (to find path)

def dfs(st, goal, graph, vis=None):
    if vis is None:
        vis = set()

    print(st)

    if st == goal:
        return True

    vis.add(st)

    for child in graph[st]:
        if child not in vis:
            if dfs(child, goal, graph, vis):
                return True

    return False
```

```python
# Static Graph
graph_static = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E', 'F'],
    'C': ['A', 'D', 'F'],
    'D': ['B', 'C', 'E', 'F'],
    'E': ['B', 'D', 'F'],
    'F': ['C', 'D', 'E', 'B']
}

print("Static Graph:", graph_static)
```

```python
# Input start & goal
start = input("Enter start node (Static): ").strip().upper()
goal = input("Enter goal node (Static): ").strip().upper()

path = dfs(start,goal,graph_static)

if path:
    print("DFS Path (Static):", path)
else:
    print("No path found")
```

```python
# Dynamic Graph 

n = int(input("Enter number of nodes: "))
e = int(input("Enter number of edges: "))
d=int(input("graph directed or not ( 1 for non directed and 0 for directed)"))
graph = {}
for i in range(n):
    node = input("Enter node name: ").strip().upper()
    graph[node] = []

for i in range(e):
    u = input("Enter starting node: ").strip().upper()
    v = input("Enter ending node: ").strip().upper()
    
    graph[u].append(v)
    if d:
        graph[v].append(u)   

print("Graph =", graph)
```

```python
# Input start & goal
start = input("Enter start node (Dyanamic): ").strip().upper()
goal = input("Enter goal node (Dynamic): ").strip().upper()

path = dfs(start,goal,graph)

if path:
    print("DFS Path (Dynamic):", path)
else:
    print("No path found")
```
